#### Sex Distribution

```python
gender_type = df['Sex'].value_counts()
plt.pie(gender_type, labels=gender_type.index,autopct='%1.0f%%')

plt.show()
```

#### Age Center

```python
x = df.index
y=df["Age"]

plt.scatter(x,y, alpha=0.4)

x_center = np.mean(x)
y_center = np.mean(y)

plt.scatter(x_center, y_center, color='red', marker='+',s=200,  label='Age center point')
```

### Missing data


```python
df["Embarked"].value_counts()
df["Embarked"].dropna()
df["Embarked"].info()

```

```python
print(df["Embarked"].unique())
df["Embarked"] = df["Embarked"].str.strip()

df["Embarked"] = df["Embarked"].str.upper()

df["Embarked"] = df["Embarked"].fillna("S")

df["Embarked"] = df["Embarked"].map({"S": 2, "C": 1, "Q": 0})
```

```python
df.head()
```

### convert Sex column to numeric

```python
df["Gender_type"] = df["Sex"].map({"male": 1, "female": 0})
df["Gender_type"]
```

```python
df.info()
```

#### Fill Age column missing data with mean

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'] = df['Age'].astype(int)
df.info()
```
